1. Import all necessary libraries and read in the data:

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import statsmodels.formula.api as smf

%matplotlib inline
````

````python
# read in data and lowercase columnnames on the fly
carseat_data = pd.read_csv('data/Carseats.csv')
carseat_data.columns = [c.lower() for c in carseat_data.columns]
carseat_data.head()
````

````python
# list of all possible explanatory continous variables
explanatory_vars = list(carseat_data.columns)
explanatory_vars = [e for e in explanatory_vars 
                        if e not in ('shelveloc', 'urban', 'us')]
explanatory_vars
````

````python
# simple linear regression with each explanatory variable
print('R squared for each possible explanatory variable:' )
for explanatory_var in explanatory_vars:
    formula = f"sales ~ {explanatory_var}"
    rs = smf.ols(formula=formula, data=carseat_data).fit().rsquared
    print(f'{explanatory_var:7s} \t \t= {rs:.3}')
# sales predicting sales included as explanatory variable for sanity check
````

````python
# best R^2 for "price", let's print the whole linear regression summary
smf.ols(formula='sales ~ price', data=carseat_data).fit().summary()
````

4a) Fit one single model with all possible explanatory variables.

````python
# Use all explanatory variables
formula = 'sales ~ compprice + income + advertising + population + price + age + education'
smf.ols(formula=formula, data=carseat_data).fit().summary()
````

4b) Interpret this model:  
How good is your model?  
Which variables seem to be important? Which do not and why? 

Our model explains 54% of the variance in price. 
Population and education are not important - they have the largest p-values

5a) Remove the less important explanatory variables. 


````python
# Remove population and education. They have the largest p-values!
model = 'sales ~ compprice + income + advertising + price + age'
smf.ols(formula=model, data=carseat_data).fit().summary()
````

5b) Is you model better now? Which metric do you look at in order to evaluate and compare your models?

Looking at the adjusted R_squared, there is not much of a difference between the two models. Since we prefer less complex over complex models, we prefer the model with less variables included.  
Overall our model doe not explain much of the variance in sales.
